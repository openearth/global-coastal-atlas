```python
import os
import rasterio
import numpy as np


```

```python
# Path to the directory containing the GeoTIFF files
directory_path = r'P:\wflow_global\hydromt\topography\copdem\elevation_masked'
output_path=r"P:\11209197-018-global-coastal-atlas\MSc_students\ClenmarRowe\Data\All_Datasets\Orig_Datasets\02_Exposure\Elevation_DEM\copDEM"

len(os.listdir(directory_path))
```

```python

n=1
# Loop over each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.tif'):
        file_path = os.path.join(directory_path, filename)
        
        # Open the GeoTIFF file
        with rasterio.open(file_path) as src:
            # Read the raster data into a NumPy array
            arr = src.read(1)
            
            # Replace values greater than 10 with -9999
            arr = np.where( (arr > 20) | (arr<0), -9999, arr)
            
            # Create a new filename by appending '_below_10m' to the original filename
            new_filename = os.path.splitext(str(filename))[0] + '_below_20m.tif'
            new_file_path = os.path.join(output_path, new_filename)
            
            # Write the modified raster data to the new GeoTIFF file
            with rasterio.open(new_file_path, 'w', **src.profile) as dst:
                dst.write(arr, 1)
                print(f"{(n/len(os.listdir(directory_path)))*100}% files edited ")
                n=n+1
```
