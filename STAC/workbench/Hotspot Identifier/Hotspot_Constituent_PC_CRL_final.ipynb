```python

directory_path=r"P:\11209197-018-global-coastal-atlas\MSc_students\ClenmarRowe\Data\Processed_data_CVI_components\Caribbean Thesis_Results\Hotspot Identification\PC"
img_directory_path=r"P:\11209197-018-global-coastal-atlas\MSc_students\ClenmarRowe\Data\Processed_data_CVI_components\Caribbean Thesis_Results\Hotspot Identification\PC\plots"
```

```python

ds=xr.open_zarr(r"P:\11209197-018-global-coastal-atlas\MSc_students\ClenmarRowe\Data\Processed_Risk_Levels_PC_FC_Merged\Caribbean_Thesis\PC_FC_RL_merged_Carib_final.zarr")
SOTC_df=ds.to_dataframe()
SOTC_gdf = gpd.GeoDataFrame(
    SOTC_df, geometry=gpd.points_from_xy(SOTC_df.lon, SOTC_df.lat), crs="EPSG:4326"
)
#Covert Bytes to string
for column in SOTC_gdf.columns:
    # Check if the column contains bytes
    if SOTC_gdf[column].dtype == 'object':
        if SOTC_gdf[column].apply(lambda x: isinstance(x, bytes)).any():
            # Decode bytes to string
            SOTC_gdf[column] = SOTC_gdf[column].apply(lambda x: x.decode() if isinstance(x, bytes) else x)
SOTC_gdf.head()
```

```python
col = [colu for colu in SOTC_gdf.columns if ("Risk_Level" in colu and "Class" not in colu)]
col

PC_col=[colu for colu in col if "PC" in colu]
PC_col
```

```python
transect=[]
max_rl=[]
for pc in PC_col:
    long_name = pc
    print_val=long_name.replace("_"," ")
    hotspot_points=pd.read_csv(os.path.join(directory_path,f"{print_val} Hotspot points.csv"))
    hotspot_points=hotspot_points.sort_values(by=long_name,ascending=False).iloc[0:1,:]
    transect.append(hotspot_points["transect_id"].values[0])
    max_rl.append(hotspot_points.iloc[:,0].values[0])
max_rl,transect,PC_col
```

```python
directory=r"P:\11209197-018-global-coastal-atlas\MSc_students\ClenmarRowe\Data\Processed_data_CVI_components\Caribbean Thesis_Results\Combination\Risk_Level_PC_builtup_pop_ns_fix"
excel_files = []

# Iterate through the files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv") or filename.endswith(".xls"):
        file_path = os.path.join(directory, filename)
        # Read the excel file
        df = pd.read_csv(file_path)
        # Append the dataframe and filename as a tuple
        excel_files.append((df))
excel_files
```

```python
a=[]
for i,excel_file in enumerate(excel_files):
    a.append(excel_file[excel_file["transect_id"]==transect[i]])
    
a[0]
```

```python
pc_df=pd.DataFrame(PC_col)
pc_df.rename(columns={pc_df.columns[0]: "risk type"}, inplace=True)
```

```python
for i,b in enumerate(a):
    if i==0:
        c=b
    else:
        c=pd.concat([c,b])
        c.iloc[i,1]=b.iloc[0,1]
c=c.iloc[:,0:13]
c["risk type"]=PC_col
c=pd.concat([pc_df,c])
c=c.iloc[5:,:]
c
```
